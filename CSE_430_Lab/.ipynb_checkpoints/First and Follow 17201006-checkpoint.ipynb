```python
################################################################
############ Author: Alim Al Razi #############################
########### University Of Asia Pacific #######################
####### Lab work-3 of Compiler Design LAB #####################
#############################################################



#re is used to import regular expression python library function 

import re

# 'A-Z' is non terminal state / Variable
#S is returning the values of states like E, R, T etc
#productions is the dictionary holding the grammer 
#first is a dict to hold first values
#follow is also a dict for keeping follow elements
# @ is used because '#' is a command variable for jupyter notebook and python 
# idx is used to keep the 'keys or index mapping of dictionary'
# for calculating follow a dollar sign is used for the non terminal Start Variable
# set () is used for preventing duplicate element if any and mapping the values in set
# a text file is used to give input

def cal_follow(s, productions, first):
    follow = set()
    #if length of the 
    if len(s)!=1 :
        return {}
    
    #dollar sign is used for the non terminal same value
    
    if(s == list(productions.keys())[0]):
        follow.add('$') 
    
    # iterating in production dictionary 
    
    for i in productions:
        for j in range(len(productions[i])):
            if(s in productions[i][j]):
               #keeping index of state in idx
                idx = productions[i][j].index(s)
                
                
                if(idx == len(productions[i][j])-1):
                    #if index value of product and current value matches break 
                    if(productions[i][j][idx] == i):
                        break
                    #else recursive call to find the follow of ith index value
                    else:
                        f = cal_follow(i, productions, first)
                        for x in f:
                            follow.add(x)
                            
                #if we are not yet at the last index
                else:
                    while(idx != len(productions[i][j]) - 1):
                        idx += 1
                        if(not productions[i][j][idx].isupper()):
                            follow.add(productions[i][j][idx])
                            break
                        #calculating first of the rightmost empty variable
                        else:
                            f = cal_first(productions[i][j][idx], productions)
                            
                            #if there is no epsilon denoted as @ add the terminal and break
                            if('@' not in f):
                                for x in f:
                                    follow.add(x)
                                break
                                
                            #if there is epsilon
                            elif('@' in f and idx != len(productions[i][j])-1):
                                f.remove('@')
                                for k in f:
                                    follow.add(k)
                                    
                            #if there is epsilon 
                            elif('@' in f and idx == len(productions[i][j])-1):
                                f.remove('@')
                                for k in f:
                                    follow.add(k)
                                
                                #recursive call the add the follows
                                f = cal_follow(i, productions, first)
                                for x in f:
                                    follow.add(x)
                            
    return follow
   
def cal_first(s, productions):
    
    first = set()
    
    for i in range(len(productions[s])):
        
        for j in range(len(productions[s][i])):
            
            c = productions[s][i][j]
            
            #if the variable is found 
            if(c.isupper()):
                f = cal_first(c, productions)
                #if no epsilon is present 
                if('@' not in f):
                    for k in f:
                        first.add(k)
                    break
                else:
                    if(j == len(productions[s][i])-1):
                        for k in f:
                            first.add(k)
                    else:
                        f.remove('@')
                        for k in f:
                            first.add(k)
                            
            #if theere is terminal values add it and break.
            else:
                first.add(c)
                break
                
    return first
                       
def main():
    productions = {}
    grammar = open("E:/4.2_All_Notes/CSE_430_Lab/grammar9.txt", "r")
    
    first = {}
    follow = {}
    
    for prod in grammar:
        l = re.split("( /->/\n/)*", prod)
        m = []
        for i in l:
            #checking the splitation and enter is pressed or -> is found
            if (i == "" or i == None or i == '\n' or i == " " or i == "-" or i == ">"):
                pass
            else:
                m.append(i)
        
        left_prod = m.pop(0)
        right_prod = []
        t = []
        
        # taking input the values after |
        for j in m:
            if(j != '|'):
                t.append(j)
            else:
                right_prod.append(t)
                t = []
        
        right_prod.append(t)
        productions[left_prod] = right_prod
    
    # s denotes the keys of the dic like E ,T ,R, F etc
    for s in productions.keys():
        first[s] = cal_first(s, productions)
    
    
    print("*****Printing the First*****")
    for lhs, rhs in first.items():
        print(lhs, ":" , rhs)
    
    print("")
    
    for lhs in productions:
        follow[lhs] = set()
    
    for s in productions.keys():
        follow[s] = cal_follow(s, productions, first)
    
    print("*****Printing the follows*****")
    for lhs, rhs in follow.items():
        print(lhs, ":" , rhs)
    
    grammar.close()

if __name__ == "__main__":
    main()
```
